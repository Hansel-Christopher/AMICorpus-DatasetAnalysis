```python
import pandas as pd

from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir('data/ami-summary/extractive') if isfile(join('data/ami-summary/extractive', f))]
fnames = []
for f in onlyfiles:
    a = f.split(".")
    fnames.append(a[0])    

def sentences_count(fname):
    lines, blanklines, sentences, words = 0, 0, 0, 0
    textf = open(fname, 'r')
    for line in textf:      
        lines += 1
        if line.startswith('\n'):
            blanklines += 1
        else:
            sentences += line.count('.') + line.count('!') + line.count('?')
            tempwords = line.split(None)
            words += len(tempwords)
    textf.close()
    return sentences

transcripts, absms, exsms = [],[],[]

for f in fnames:
    transcripts.append(sentences_count('data/ami-transcripts/'+f+'.transcript.txt'))
    absms.append(sentences_count('data/ami-summary/abstractive/'+f+'.abssumm.txt'))
    exsms.append(sentences_count('data/ami-summary/extractive/'+f+'.extsumm.txt'))
    print ("Transcript Sentences  : ", sentences_count('data/ami-transcripts/'+f+'.transcript.txt'))
    print ("Abstractive summary Sentences  : ", sentences_count('data/ami-summary/abstractive/'+f+'.abssumm.txt'))
    print ("Extractive summary Sentences  : ", sentences_count('data/ami-summary/extractive/'+f+'.extsumm.txt'))
    
data = {'Transcripts': transcripts,
        'Abstractive_summary_sentences': absms,
        'Extractive_summary_sentences' : exsms
       }
  
df = pd.DataFrame(data,columns=['Transcripts', 'Abstractive_summary_sentences','Extractive_summary_sentences'], index=fnames)
df1=df.head(10)
df1.plot.bar(rot=0,ylim=(0,100))
```
